```python
import pandas as pd
```

```python
train = pd.read_csv("../data/bike/train.csv", parse_dates=["datetime"])
test = pd.read_csv("../data/bike/test.csv", parse_dates=["datetime"])
```

```python
#sns.countplot(data=train, x="windspeed", hue=)
```

## 날짜를 정수값으로 각 컬럼별로 분리하기

```python
train["year"] = train["datetime"].dt.year
train["month"] = train["datetime"].dt.month
train["day"] = train["datetime"].dt.day
train["hour"] = train["datetime"].dt.hour
```

```python
test["year"] = test["datetime"].dt.year
test["month"] = test["datetime"].dt.month
test["day"] = test["datetime"].dt.day
test["hour"] = test["datetime"].dt.hour
```

## windspeed 값 중에 0 을 평균값으로 채워넣기

```python
mean_windspeed = train["windspeed"].mean()
train.loc[train["windspeed"] == 0, "windspeed"] = mean_windspeed
```

```python
mean_windspeed = test["windspeed"].mean()
test.loc[test["windspeed"] == 0, "windspeed"] = mean_windspeed
```

## windspeed, temp, atemp 반올림

```python
train["windspeed"] = train["windspeed"].round()
train["temp"] = train["temp"].round()
train["atemp"] = train["atemp"].round()

test["windspeed"] = test["windspeed"].round()
test["temp"] = test["temp"].round()
test["atemp"] = test["atemp"].round()
```

## season, weather 컬럼을 one hot encoding

```python
train_season = pd.get_dummies(train["season"], prefix="season")
train_weather = pd.get_dummies(train["weather"], prefix="weather")
test_season = pd.get_dummies(test["season"], prefix="season")
test_weather = pd.get_dummies(test["weather"], prefix="weather")


train = pd.concat([train, train_season], axis=1)
train = pd.concat([train, train_weather], axis=1)
test = pd.concat([test, test_season], axis=1)
test = pd.concat([test, test_weather], axis=1)
```

## 날짜
- Holiday = 1, WorkingDay = 0 ---> 0
- Holiday = 0, WorkingDay = 1 ---> 0
- Holiday = 0, WorkingDay = 0 ---> 1

```python
train.loc[(train["holiday"] == 1) | (train["workingday"] == 1), "dd"] = 0
train.loc[(train["holiday"] == 0) & (train["workingday"] == 0), "dd"] = 1

test.loc[(test["holiday"] == 1) | (test["workingday"] == 1), "dd"] = 0
test.loc[(test["holiday"] == 0) & (test["workingday"] == 0), "dd"] = 1
```

```python
def rmsle(predict, actual):
    """
    predict : 예측값
    actual : 실제값
    """
    predict = np.array(predict)
    actual = np.array(actual)
    
    log_predict = np.log(predict + 1)
    log_actual = np.log(actual + 1)
    
    distance = log_predict - log_actual
    
    squared_distance = distance ** 2
    mean_square_distance = squared_distance.mean()
    score = np.sqrt(mean_square_distance)
    
    return score
```
